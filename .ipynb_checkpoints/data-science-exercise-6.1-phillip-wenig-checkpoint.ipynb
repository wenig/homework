```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Download data set

```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv', 
    index_col='id'
                )
```

```python
df
```

## Drop complete empty rows

Empty rows do not deliver any insights in a dataset

```python
df = df[~(pd.isna(df.full_name) & 
          pd.isna(df.gender) &
          pd.isna(df.first_name) &
          pd.isna(df.email) &
          pd.isna(df.age))]
df
```

## Clean gender column

I don't think we should delete the data with id `6` because maybe this person is beyond the male-female classification (transgender etc). Instead I'll create a new category `unknown`.

```python
df.gender.unique()
```

```python
df.loc[pd.isna(df.gender), 'gender'] = 'unknown'
```

```python
df
```

## Clean age column

Row with id `19` probably meant 78 and typed the minus sign unintentionally.

Row with id `10` is a bit trickier, since we don't know how important the age column is. I'd recommend to look at the age distribution of `Male` samples to set an approximated value.

```python
df[~df.age.apply(lambda x: x.isdigit())]
```

```python
df.loc[19, 'age'] = df.loc[19, 'age'][1:]
```

```python
df[df.age.apply(lambda x: x.isdigit()) & (df.gender == 'Male')].age.astype(int).plot(kind='hist')
```

Here, we nicely can see the split in ages and `old` is probably meant to be over 60. Therefore, we set the value of id `10` to the average of the 'old' samples.

```python
df.loc[10, 'age'] = df[df.age.apply(lambda x: x.isdigit() and int(x) >= 60)].age.astype(int).mean()
```

```python
df.loc[:, 'age'] = df.age.astype(int)
```

```python
df
```

## Clean email column

This column needs to be unique. Rows with id `15` and `16` have the same email. Moreover, everything else is also the same. Hence we delete row `16`.

```python
df[df.email.duplicated()]
```

```python
df = df[~df.email.duplicated()]
```

Row `21` has no email at all. As we don't know whether we need it, I wouldn't delete this sample and leave this field empty.

## Index

There is one row without id left. This row gets an index assigned. We simply add an 1 to the last id.

```python
df = df.reset_index()
```

```python
df
```

```python
df.loc[pd.isna(df.id), 'id'] = df.id.max() + 1
```

```python
df = df.set_index('id')
df
```
